```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/spam')
df
```

```python
df.Category.value_counts()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Message'], df['Category'], test_size=0.2, random_state=42)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

# count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
```

# Example

```python
emails=[
    'Upto 20% discout reward',
    'Hey Sunny discount upto'
]
emails_cv=v.fit_transform(emails)
emails_cv.toarray()
```

```python
v.get_feature_names_out()
```

```python
pd.DataFrame(emails_cv.toarray(),columns=v.get_feature_names_out())
```

```python
v=CountVectorizer()
X_train_cv=v.fit_transform(X_train.values)
X_train_cv.toarray()
```

```python
X_train_cv.shape
```

```python
v.get_feature_names_out()
```

```python
pd.DataFrame(X_train_cv.toarray(),columns=v.get_feature_names_out())
```

```python
v.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
v_text_cv=v.transform(X_test)
```

```python
v_text_cv.toarray().shape
```

```python
model.score(v_text_cv,y_test)
```

```python
from sklearn.metrics import classification_report

y_pred=model.predict(v_text_cv)
print(classification_report(y_test,y_pred))
```

```python
emails=[
    'Upto 20% discout on parking, exclusive offer just for you. Dont miss this reward!',
    'Hey Sunny ,can we get together to watch footbal game tomorrow?'
]
emails_cv=v.transform(emails)
model.predict(emails_cv)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Print the classification report for the Multinomial Naive Bayes model
print(classification_report(y_test,y_pred))

# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['ham', 'spam'], yticklabels=['ham', 'spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Multinomial Naive Bayes')
plt.show()
```
